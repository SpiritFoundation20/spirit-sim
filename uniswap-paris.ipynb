```python
import pandas as pd

from uniswap import Uniswap

u = Uniswap()
u.url
```

```python
uniswap_pairs = u.get_pair_mapping(exchange='Uniswap')

uniswap_pairs['Symbol'] = uniswap_pairs['fsym'] + '/' + uniswap_pairs['tsym']
```

```python
uniswap_pairs.set_index('Symbol').head()
```

```python
u.get_ohlcv('ADAI/ETH', '1d', 30)['Volume'].mean()
```

```python
uniswap_pairs['Volume'] = uniswap_pairs['Symbol'].apply(lambda x: u.get_ohlcv(x,'1d',30)['Volume'].mean())
```

```python
uniswap_pairs[['Symbol','Volume']].to_csv('Uniswappairs.csv',index=False)
```
